```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', version=1, cache=True)
X = mnist.data
y = mnist.target
```

```python
X.shape
```

```python
data_5000 = X[0:5000,:]
labels_5000 = y[0:5000]
```

```python
data_5000= StandardScaler().fit_transform(data_5000)
```

```python
model = TSNE(n_components=2, random_state=0)
```

```python
tsne_data = model.fit_transform(data_5000)
```

```python
data_5000
```

```python
tsne_data.shape
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x = tsne_data[:,0], y = tsne_data[:,1], hue = labels_5000, palette = sns.hls_palette(10), legend = 'full');
```
